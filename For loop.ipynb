# For loop

```python
integers = [1,2,3,4,5]
```

```python
for number in integers:
    print(number)
```

```python
for number in integers:
    print(number + number)
```

```python
for number in integers:
    print('whole number')
```

```python
dictionary ={'Name':'Ronak shaw','Age':23,'Gender':'Male'}
for dict_ in dictionary.values():
    print(dict_)
```

```python
dictionary ={'Name':'Ronak shaw','Age':23,'Gender':'Male'}
for dict_ in dictionary.keys():
    print(dict_)
```

```python
dictionary ={'Name':'Ronak shaw','Age':23,'Gender':'Male'}
for key,value in dictionary.items():
    print(key,'->',value)
```

```python
#nested loop
regions = ['North','South']
products = ['Fabric','Yarn','Dye']
for region in regions:
    for product in products:
        print('Analyzing', product, 'sales in', region)
```
