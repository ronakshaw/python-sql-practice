# Automatic file sorter in file explorer

```python
import os,shutil
```

```python
path = r"/Users/ronak/Documents/Python project/"
```

```python
file_name = os.listdir(path)
```

```python
folder_names = ['txt files', 'png file', 'pdf file', 'xlsx file']

for loop in range(0,4):
    if not os.path.exists(path + folder_names[loop]):
        os.makedirs(path + folder_names[loop])

for file in file_name:
    if ".txt" in file and not os.path.exists(path + "txt files/" + file):
        shutil.move(path + file, path + "txt files/" + file)
    elif ".png" in file and not os.path.exists(path + "png file/" + file):
        shutil.move(path + file, path + "png file/" + file)
    elif ".pdf" in file and not os.path.exists(path + "pdf file/" + file):
        shutil.move(path + file, path + "pdf file/" + file)
    elif ".xlsx" in file and not os.path.exists(path + "xlsx file/" + file):
        shutil.move(path + file, path + "xlsx file/" + file)
```
