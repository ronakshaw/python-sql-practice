```python
from bs4 import BeautifulSoup
import requests
```

```python
my_headers = {
    'User-Agent': 'IndianITCompaniesScraper (contact: rkronak06@email.com)'
}
url = 'https://en.wikipedia.org/wiki/List_of_Indian_IT_companies'
page = requests.get(url, headers= my_headers)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
table = soup.find_all('table')[0]
```

```python
titles = table.find_all('th')
```

```python
table_titles = [header.text.strip()for header in titles]
print(table_titles)
```

```python
column_data = table.find_all('tr')
```

```python
import pandas as pd
import numpy as np

all_team_rows = []

for rows in column_data:
    row_data = rows.find_all('td')
    indv_row = [data.text.strip() for data in row_data]
    
    if indv_row:
        # This is the crucial fix you are missing:
        # If the row has only 1 value (the city), push it to the 5th column
        if len(indv_row) == 1:
            indv_row = ['', '', '', '', indv_row[0]]
            
        all_team_rows.append(indv_row)

# 1. Build the Dataframe
df = pd.DataFrame(all_team_rows, columns=table_titles)

# 2. Turn empty spaces into NaN so pandas can see them
df = df.replace('', np.nan)

# 3. Forward fill the company metadata down
columns_to_fill = ['Name', 'Headquarters', 'Total income', 'Headcount']
df[columns_to_fill] = df[columns_to_fill].ffill()

# Display the output
df
```
