```python
# nested if
attendance = int(input('Enter attendance %')) 
marks = int(input('Enter marks')) 

if attendance >= 75:
    if marks >= 90:
        print('Grade A')
    elif marks >= 80:
        print('Grade B')
    elif marks >= 70:
        print('Grade C')
    elif marks >= 60:
        print('Grade D')
    else:
        print('fail')
else:
    print('low attendance')
```
