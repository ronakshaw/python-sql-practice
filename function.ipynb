# Function

```python
def first_function():
    print('did it')
```

```python
first_function()
```

```python
def squared_num(number):
    print(number**2)
```

```python
squared_num(3)
```

```python
def squared_num_custom(number,power): # (what we write inside this bracket is called argument)
    print(number**power)
```

```python
squared_num_custom(3,6)
```

```python
#arbitrary argument (when we don't know how many arguments we're gonna take)
def number_args(*number):
    print (number[0]*number[1])
```

```python
number_args(7,9,4,5,6,7,0)
```

```python
tuple = (7,9,4,5,6,7,0)
def number_args(*number):
    print (number[0]*number[1])
```

```python
number_args(*tuple)
```

```python
#keyword argument
def squared_num_custom(number,power):
    print(number**power)
```

```python
squared_num_custom(power=6,number=4)
```

```python
#arbitrary keyword argument(means we don't knoe how many keyword argument we need to pass)
def phone_num(**number):
    print('my number is: ' + number['integer'] + ' My alternate number is: ' +number['integer2'])
```

```python
phone_num(integer = '7865' ,integer2 = '4567')
```
