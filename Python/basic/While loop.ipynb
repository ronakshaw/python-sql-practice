# while loop

```python
number = 0
while number <5:
    print(number)
    number = number+1
    
```

```python
number = 0
while number <5:
    print(number)
    if number==3:
        break
    number = number+1
```

```python
number = 0
while number <5:
    print(number)
    if number==6:
        break
    number = number+1
else:
    print('no longer <5')
```
