```python
set_1 ={1,23,3,5,3,4,5,2,9}
print(set_1)
set_2 = {6,3,7,6,8,7,56,7,9,0}
print(set_2)
print(set_1|set_2) # combines unique values of both sets
print(set_1^set_2) # shows if a value is present one or other but not in both
print(set_1&set_2) # shows values that are present in both
print(set_1-set_2) # removes all the values of set_2 which is present in set_1
```

```python
dictionary = {      # key:value pair
    'Name': 'Ronak',
    'Age': 22,
    'Gender': 'Male',
    'Favourite_ice_cream': ['chocolate','vanilla']
}
print(dictionary.values())
print(dictionary.keys())
print(dictionary.items())
print(dictionary['Name']=='Tithi')
```

```python
dictionary = {      # key:value pair
    'Name': 'Ronak',
    'Age': 22,
    'Gender': 'Male',
    'Favourite_ice_cream': ['chocolate','vanilla']
}
dictionary['Name']='Tithi'
dictionary['Gender']='Female'
del dictionary['Age']
print(dictionary.items())
```
